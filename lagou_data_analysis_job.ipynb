## 创建写入MySQL数据库的函数insert_mysql

```python
def insert_mysql(id, company_name, industry_field, finance_stage, company_size,
                 position, city, salary, experience, education, skill, welfare):
    """
        将数据写入表data_analysis中
    """
    try:
        cursor = db.cursor()
        sql = 'insert into data_analysis(' \
              'id, company_name, industry_field, finance_stage, company_size,' \
              'position, city, salary, experience, education, skill, welfare) ' \
              'values ("%d", "%s", "%s", "%s", "%s", "%s", "%s", "%s", "%s", "%s", "%s", "%s")' \
              % (id, company_name, industry_field, finance_stage, company_size, 
                 position, city, salary, experience, education, skill, welfare)
        print(sql)
        cursor.execute(sql)
        print(cursor.lastrowid)
        db.commit()
    except Exception as e:
        print(e)
        db.rollback()
```

## 创建爬取数据的函数get_data

```python
def get_data(url, headers):
    """
        获取拉勾网“数据分析”相关职位数据
    """
    for i in range(1, 31):
        # 爬取若干页需要延迟线程运行，否则会报错
        if i % 3 == 0:
            time.sleep(random.randint(15, 20))
        print('第' + str(i) + '页')
        time.sleep(random.randint(15, 20))
        if i == 1:
            data = {'first': 'true', 'pn': 1, 'kd': '数据分析'}
        else:
            data = {'first': 'false', 'pn': i, 'kd': '数据分析'            }
        response = requests.post(url=url, data=data, headers=headers)
        result = json.loads(response.text)              
        job_info = result['content']['positionResult']['result']
        for job in job_info:
            global count
            count += 1
            # 公司名称
            company_name = job['companyShortName']
            # 行业领域
            industry_field = job['industryField']
            # 融资阶段
            finance_stage = job['financeStage']
            # 公司规模
            company_size = job['companySize']
            # 岗位名称
            position = job['positionName']
            # 岗位地点
            city = job['city']
            # 岗位薪水
            salary = job['salary']
            # 岗位经验要求
            experience = job['workYear']
            # 岗位学历要求
            education = job['education']
            # 工作技能要求
            if len(job['skillLables']) > 0:
                skill = ','.join(job['skillLables'])
            else:
                skill = 'None'
            # 工作福利
            welfare = job['positionAdvantage']
            # 写入数据库lagou_job中data_analysis表
            insert_mysql(count, company_name, industry_field, finance_stage, company_size, 
                         position, city, salary, experience, education, skill, welfare)
        print('已完成第'+ str(i) + '页')
        # 爬取若干页需要延迟线程运行，否则会报错
        if i % 7 == 0:
            time.sleep(random.randint(15, 20))
```

## 数据分析岗工作技能分析

### 工作技能矩形树图

```python
from pyecharts.charts import TreeMap
```

```python
# 由于表中的skill列中每行可能包含不止一项技能，因此需要对每行进行拆分
skill_list = []
for i in df['skill']:
    skill = i.split(',')
    for j in range(len(skill)):
        if skill[j] in skill_list:
            continue
        else:
            if skill[j] != 'None':
                skill_list.append(skill[j])
```

```python
# 统计每项技能出现的次数
skill_count = []
for item in skill_list:
    num = 0
    for i in df['skill']:
        skill = i.split(',')
        for j in range(len(skill)):
            if skill[j] == item:
                num += 1
            else:
                continue
    skill_count.append(num)
```

```python
def skill_info():
    """
        将技能名称和对应的次数合并，方便在矩形树图中展示信息
    """
    for i in range(len(skill_count)):
        data = {}
        data['name'] = skill_list[i] + ' ' + str(skill_count[i])
        data['value'] = skill_count[i]
        yield data
```

# 统计推断

为了进一步探究影响薪资的变量，我们建立多元回归模型、对数线性回归模型和随机森林回归模型来对薪资进行预测。

## 多元线性回归模型：以城市、工作经验、学历要求为自变量

### 对城市、工作经验、学历均使用标签编码

```python
from sklearn.preprocessing import LabelEncoder as LE
from sklearn.linear_model import LinearRegression as LR
```

```python
# 将各字符分类变量重编码为数值分类变量
le = LE()   
# 城市重编码                                           
df['city'] = le.fit_transform(df['city'].values)
# 学历重编码 
df['education']=df['education'].replace('不限',0)                  
df['education']=df['education'].replace('大专',1)
df['education']=df['education'].replace('本科',2)
df['education']=df['education'].replace('硕士',3)
# 工作经验重编码
df['experience']=df['experience'].replace('不限',0)
df['experience']=df['experience'].replace('应届毕业生',1)
df['experience']=df['experience'].replace('1年以下',2)
df['experience']=df['experience'].replace('1-3年',3)
df['experience']=df['experience'].replace('3-5年',4)
df['experience']=df['experience'].replace('5-10年',5)
df['experience']=df['experience'].replace('10年以上',6)
```

```python
from sklearn.model_selection import train_test_split

# 自变量
df_lm1 = df[['city','experience','education']]
X = df_lm1
# 因变量
y = df['salary_avg']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=88)

lin_reg1 = LR()
lin_reg1.fit(X_train,y_train)
```

```python
print('截距：', lin_reg1.intercept_)
print('系数：', lin_reg1.coef_)
```

```python
# 构建预测函数
def salary_prediction(city,exper,edu):
    city_index = list(('北京','上海','广州','深圳','杭州','成都','武汉','南京','厦门','重庆','长沙',
                       '佛山','太原','宁波','东莞','泉州','苏州','西安','贵阳')).index(city)
    edu_index = list(('不限','大专','本科','硕士')).index(edu)
    salary_pred = lin_reg1.predict([[city_index,exper,edu_index]])
    print('地点：', city)
    print('工作经验：', exper)
    print('学历：', edu)
    print('月薪：','%.2f'%salary_pred, '元/月')
```

```python
# 预测地点在厦门、工作经验要求为应届毕业生、学历要求为本科的薪资
salary_prediction('厦门',1,'本科') # 1表示应届毕业生
```

```python
salary_prediction('厦门',1,'硕士') # 1表示应届毕业生
```

### 将城市的编码转化成独热编码

工作经验的数字越大可以表示经验越丰富，学历的数字越大可以表示学历越高。而城市属性没有大小差异，因此需要转化成独热编码。

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([('city', OneHotEncoder(), [0])], remainder = 'passthrough')

df_lm2 = df[['city','experience','education']]
data_hot = ct.fit_transform(df_lm2)
X = data_hot.toarray()
```

```python
y = df['salary_avg']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=88)

lin_reg2 = LR()
lin_reg2.fit(X_train,y_train)
```

查看前5个数据的预测结果

```python
some_data = X[:5]
some_label = y[:5]
print('Predictions:\t', lin_reg2.predict(some_data))
print('Labels:\t', list(some_label))
```

均方误差

```python
from sklearn.metrics import mean_squared_error

y_pred = lin_reg2.predict(X_train)
mse = mean_squared_error(y_train, y_pred)
rmse = np.sqrt(mse)
print('RMSE =',rmse)
```

使用交叉验证的方法测试模型的误差

```python
from sklearn.model_selection import cross_val_score

# 10折交叉验证
scores_lin_reg = cross_val_score(lin_reg2, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
rmse_scores_lin_reg = np.sqrt(-scores_lin_reg)
print(rmse_scores_lin_reg)
print('RMSE均值：',rmse_scores_lin_reg.mean())
print('RMSE标准差：',rmse_scores_lin_reg.std())
```

## 对数线性回归模型：对薪资取对数，以城市、工作经验、学历要求为自变量

为了减少异常值的影响，对平均薪资取对数。

### 对数线性回归

```python
import math

df['log_salary'] = [math.log(s) for s in df['salary_avg']]
```

```python
df_lm3 = df[['city','experience','education']]
data_hot = ct.fit_transform(df_lm3)
X = data_hot.toarray()
```

```python
y = df['log_salary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=88)

lin_reg3 = LR()
lin_reg3.fit(X_train,y_train)
```

均方误差

```python
y_pred = lin_reg3.predict(X_train)
mse = mean_squared_error(y_train, y_pred)
rmse = np.sqrt(mse)
print('RMSE =', rmse)
```

使用交叉验证的方法测试模型的误差

```python
scores_lin_reg = cross_val_score(lin_reg3, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
rmse_scores_lin_reg = np.sqrt(-scores_lin_reg)
print(rmse_scores_lin_reg)
print('RMSE均值：',rmse_scores_lin_reg.mean())
print('RMSE标准差：',rmse_scores_lin_reg.std())
```

### 随机森林回归

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(random_state=88)
forest_reg.fit(X_train, y_train)
```

```python
scores_forest_reg = cross_val_score(forest_reg, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
rmse_scores_forest_reg = np.sqrt(-scores_forest_reg)
print(rmse_scores_forest_reg)
print('RMSE均值：',rmse_scores_forest_reg.mean())
print('RMSE标准差：',rmse_scores_forest_reg.std())
```

GridSearchCV调整模型的超参数

```python
from sklearn.model_selection import GridSearchCV
```

```python
"""
    n_estimators太小，容易欠拟合
    n_estimators太大，计算量太大，且n_estimators到一定的数量后，再增大n_estimators获得的模型提升会很小
"""
param_grid = [
    {'n_estimators': [3, 30, 40, 80], 'max_features': [2, 4, 6]},
    {'bootstrap': [False], 'n_estimators': [3, 40], 'max_features': [2, 4, 6]},
]

forest_reg = RandomForestRegressor(random_state=88)
grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error')
```

```python
grid_search.fit(X_train, y_train)
```

最好的参数组合

```python
grid_search.best_params_
```

查看所有测试参数组合的情况

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres['mean_test_score'], cvres['params']):
    print(np.sqrt(-mean_score), params)
```

使用测试集检验RMSE

```python
final_model_forest_reg = grid_search.best_estimator_
final_forest_reg_predictions = final_model_forest_reg.predict(X_test)
final_forest_reg_mse = mean_squared_error(y_test, final_forest_reg_predictions)
final_forest_reg_rmse = np.sqrt(final_forest_reg_mse)
print('测试集在随机森林回归模型中RMSE的均值：',final_forest_reg_rmse)
```

```python
final_model_lin_reg = lin_reg3
final_lin_reg_predictions = final_model_lin_reg.predict(X_test)
final_lin_reg_mse = mean_squared_error(y_test, final_lin_reg_predictions)
final_lin_reg_rmse = np.sqrt(final_lin_reg_mse)
print('测试集在对数线性回归模型中RMSE的均值：',final_lin_reg_rmse)
```

可以看到，随机森林回归模型的误差比对数线性回归模型的误差小。

## 对数线性回归模型：加入部分工作技能要求

除了城市、工作经验、学历要求对薪资有影响外，部分工作技能要求对薪资也有一定影响。因此，我们在前面模型的基础上继续改善模型，通过加入部分工作技能的虚拟变量，检验不同工作技能对薪资的影响。

根据 <span><a href="#工作技能矩形树图" data-toc-modified-id="工作技能矩形树图-4.4.1"><span class="toc-item-num">4.4.1&nbsp;&nbsp;</span>工作技能矩形树图</a></span> ，我们挑选了20个工作技能要求，其中前10个为软件方面的要求，后10个为业务技能方面的要求。

```python
skill_name = ['SQL','MySQL','SQLServer','BI','Python','Hive','Hadoop','Spark','SPSS','Java',
              '数据分析','数据运营','数据挖掘','数据处理','数据库','数据仓库','可视化','ETL','用户研究','行业分析']
```

```python
skills = [item for item in df['skill']]

# 判断skills中每个元素是否含有skill_name中的技能
for i in range(len(skill_name)):
    tf_list = []
    for j in range(len(skills)):
        tf = skill_name[i] in skills[j]
        tf_list.append(int(tf))
    df[skill_name[i]] = tf_list 
```
